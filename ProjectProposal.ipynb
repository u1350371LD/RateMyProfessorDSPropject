Complete a stakeholder analysis for your project.
Who may be affected by your project and its outcomes? How could you project be used for harm?
“There are no ethical considerations” must be strongly defended. No one successfully done this before in this class.**


RateMyProfessor as a platform is inherently prone to ethical implications regarding reputation, and this project has the potential to 
exacerbate those issues by scaling them up from the level of individuals to that of instituions and fields of study. The subjects of the RateMyProfessor website have inflicted on them the same consequences that yelp has on restaurants. In the same way that a slew of bad reviews (warrented or not) of an applebees can hurt it's reputation to the point of closure, the reviews on RateMyProfessor can result in series concequences for the carreer of instructors. The effect of our project could then bring this same issue to entire academic institutions and those within them. It is possible that certain departments in public universites could lose funding if our project shows unpopularity for a certain field. A prospective student might find value in seeing our results, and make decisions about enrollment based on the amalgamated data of those who came before them. This, however, risks perpetuating underlying societal biases by conflating popularity with institutional quality. The way we present our data should be very clear about this distinction to avoid reputational harm.

How are you planning to analyze your data?

Aggregation and Grouping

We will group professors by department or degree category and calculate summary statistics such as:

- Average star rating
- Average difficulty score
- Average take-again percentage
- Average tag frequencies

This will allow us to compare departments and identify which programs tend to have higher difficulty levels or lower student satisfaction.

Correlation Analysis

We will compute correlations between variables such as:

- Difficulty vs. star rating
- Difficulty vs. take-again percentage
- Tags vs. difficulty ratings

This will help determine whether perceived difficulty strongly affects professor ratings.

Ranking Analysis

We may create rankings of departments based on metrics such as:

- Highest average difficulty
- Lowest average rating
- Highest frequency of “tough grader” or “lots of homework” tags

Comparative Analysis

Finally, we will compare degree categories such as:

- Engineering
- Science
- Humanities
- Social sciences

This will help us determine whether certain academic areas consistently report more difficult classroom experiences.

Through these methods we hope to determine whether fields like engineering truly have the hardest professors, or if other majors report similar or higher levels of difficulty.
