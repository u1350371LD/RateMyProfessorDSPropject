**Background and Motivation: 
Discuss your motivations and reasons for choosing this project, especially any background or research interests that may have
influenced your decision.**

As students, we are very concerned with the quality of instructors that teach us. Many students often switch professors due to factors like teaching style, reputation, class structure, and more. The quality of instructor often means more to a student than the class difficulty. I (Luke Dunbar) personally know someone who dropped an entire major because of a particular professor. As students, it is vital that we recieve quality instruction to actually gain benefit from our degrees so we are properly prepared for the workplace.

RateMyProfessor (https://www.ratemyprofessors.com) is a website where students can grade/rate their instructors and provide feedback. While these metrics can be biased, they usualy paint a fairly accurate description of the professor. This project can help students and upper faculty determine which professors or even majors have the most effective, easiest, or worst time. It can help us make more informed decisions on who we trust to impart career defining knowledge.

On a more personal level. I'm (Luke Dunbar) curious in which degrees actually have it hardest. As a Mechanical Engineer, my classmates and I like to believe that we have the hardest classes, so, I'd like to see if it's actually true.

Primary Question: Which degree paths suffer the most?

The goal of this project is to determine which degrees have the toughest/ worst professors. On a larger scale, we want to apply the data handling and visualiziation techniques taught in this class. There is no single metric that can accurately determine how difficult or frustrating a degree path will be, but by using the "raw" data and interpretation, we can paint a picture of how difficult courses are based on the entire class experience, not just if the content is difficult. 

From this project we hope to benefit in the data handling sphere and in the data interpretation sphere. As mentioned, this data set will allow us to practice many of the data wrangling skills that we learned in class, however, because this data exists in bias, context, and non numerical forms, it is also important that we practice intensive interpretation, mindfullness, and awareness. Not only can we hope to learn about what to avoid when scheduling, but a lot about quantifying different data types, interpreting results, and context.
